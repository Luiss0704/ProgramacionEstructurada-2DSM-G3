# Programa de calificaciones

Este programa calcula el promedio de tres calificaciones y evalúa si un alumno **aprobó** o **reprobó**

## Definición de variables

```python
calif1 = 0.0
calif2 = 0.0
calif3 = 0.0
promedio = 0.0
```

## Entrada de datos

```python
calif1 = float(input("Ingrese la primera calificación:"))
calif2 = float(input("Ingrese la segunda calificación:"))
calif3 = float(input("Ingrese la tercera calificación:"))
```

## Proceso: Cálculo de promedio

```python
promedio = (calif1 + calif2 + calif3) / 3
```

## Comparación con una Estructura selectiva doble

```python
if promedio >= 7:
    print("APROBADO")
else:
    print("REPROBADO")
```
